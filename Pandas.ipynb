```python
drinks=pd.read_csv("https://raw.githubusercontent.com/nunnarilabs/ml/master/pandas/drinks.csv")
drinks.head()
```

```python
drinks.info()
```

```python
drinks.beer_servings.mean()
```

```python
drinks.beer_servings.max()
```

```python
drinks.beer_servings.min()
```

```python
drinks[drinks.continent=='Asia'].describe()
```

```python
#aggregation function
drinks[drinks.continent=='Asia'].beer_servings.agg(['mean','min','max'])
```

```python
#filtering that is grouping based on particular data
drinks.groupby('continent').beer_servings.max()
```

```python
drinks.groupby('continent').beer_servings.agg(['mean','min','max'])
```
